### Ejercicio 2

**Lea estos ficheros desde un programa Python, y realice las labores de pre-procesamiento habituales: división en frases, división en palabras y conversión a las formas normales. Utilice los frameworks que prefiera, se sugiere NLTK o spaCy**

* División en frases:

__nltk__

```python
import nltk 

sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')

sents = sent_tokenizer.tokenize(engl)
for sent in sents:
    print(sent)
```

__spacy__

```python
# Importamos las librerías que corresponda
# Hay que añadir los modelos que se utilicen. Previamente cargados con python -m spacy download es_core_news_md

import spacy
from spacy import displacy
import pandas as pd


nlp_es = spacy.load('es_core_news_md')
nlp_eng = spacy.load('en_core_web_sm')
```

```python
# Ejecutamos la pipeline de análisis

docum_analizado = nlp_es(espa)
docum_analyzed = nlp_eng(engl)
```

```python
# Imprimimos frases

for sent in docum_analizado.sents:
    print(sent.text)
for sent_eng in docum_analyzed.sents:
    print(sent_eng.text)
```

* División en palabras:

__nltk__

```python
print("Tokens")
tokens = nltk.word_tokenize(engl)
print(tokens[:15])
```

```python
# Imprimimos tokens

# POS tagging with Spacy 
spacy_pos_tagged = [(palabra.text) for palabra in docum_analizado]
pd.DataFrame(spacy_pos_tagged, columns=['Palabra']).T
```

```python
# Ahora en ingles
spacy_eng_pos_tg = [(word.text) for word in docum_analyzed]
pd.DataFrame(spacy_eng_pos_tg, columns=['Palabra']).T
```

* Conversión a las formas normales:

__spacy__

```python
# Imprimimos formas canónicas

# POS tagging with Spacy 
spacy_pos_tagged = [(palabra.lemma_) for palabra in docum_analizado]
pd.DataFrame(spacy_pos_tagged, columns=['F. Canónica']).T
```

```python
# Ahora en ingles
spacy_eng_pos_tg = [(word.lemma_) for word in docum_analyzed]
pd.DataFrame(spacy_eng_pos_tg, columns=['F_Canónica']).T
```

**Comente las diferencias entre los resultados obtenidos en español y en inglés. ¿Hay diferencias entre utilizar NLTK y spaCy?**

La librería NLTK sirve mucho pre-procesamiento, crear los tokens, stemming, Part Of Speech, etc. SpaCy puede hacer muchísimas cosas al estilo de NLTK pero es bastante más rápido.

La diferencia básica entre las dos bibliotecas es el hecho de que NLTK contiene una amplia variedad de algoritmos para resolver un problema, mientras que spaCy contiene solo uno, pero el mejor algoritmo para resolver un problema.

NLTK se lanzó en 2001, mientras que spaCy es relativamente nuevo y se desarrolló en 2015. SpaCy es una biblioteca para el procesamiento avanzado de lenguaje natural en Python. Incluye modelos estadísticos pre-entrenados y vectores de palabras, admite tokenización para más de 45 idiomas. Provee etiquetado, análisis y reconocimiento de entidades nombradas y una fácil integración de aprendizaje profundo. Es muy sencilla de utilizar, basta con incorporar desde un programa Python el modelo para el español.

Devuelve un etiquetado POS completo, no sólo indica la función de la palabra en la oración, sino otros datos tales como tiempo verbal, persona, número, modo, género,

En conclusión, tanto NLTK como spaCy ofrecen opciones muy buenas para construir un sistema NLP. Sin embargo, como hemos visto, spaCy es la herramienta adecuada para usar en un entorno de producción, su lema - _proporcionar un servicio en lugar de ser una herramienta_ - explica de su extrema facilidad de uso y rendimiento.
